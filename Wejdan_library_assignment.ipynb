```python
# Python Library Management System

# Install required libraries (This should already exist in colab)
!pip install requests

import pickle
import sqlite3
import requests

# 1. Data Types and Variables
book_title = "Python Programming"
book_quantity = 5
book_price = 29.99
is_available = True

print(f"Book: {book_title}, Quantity: {book_quantity}, Price: ${book_price}, Available: {is_available}")

# 2. Lists
books = ["Python Basics", "Advanced Python", "Data Science with Python"]

# TODO: Add two more books to the list
books.append("Clean Code")
books.insert(1, "Fluent Python")

print("Updated book list:", books)

# 3. Loops and range()
print("\nAvailable books:")
for i in range(len(books)):
    print(f"{i}: {books[i]}")

# 4. User Input
new_book = input("Enter a new book title: ")
books.append(new_book)
print("Updated book list:", books)

# 5. Functions and Conditional Statements
def check_availability(book_name):
    # Check if the book is in the list and return True if available, False otherwise
    if book_name in books:
        return True
    else:
        return False
    pass

# Test the function
print(f"Is 'Python Basics' available? {check_availability('Python Basics')}")

# 6. File I/O
def save_books_to_file(filename):
    with open(filename, 'w') as f:
        for book in books:
            f.write(f"{book}\n")


def read_books_from_file(filename):
    # Open the file, read each line, and strip the trailing newline character
    try:
        with open(filename, 'r') as f:
            return [line.strip() for line in f]
    except FileNotFoundError:
        print(f"Error: {filename} not found.")
        return []

# Test file I/O
save_books_to_file("books.txt")
!cat books.txt  # Display file contents in Colab

# Test reading the books back from the file
loaded_books = read_books_from_file("books.txt")
print("Books loaded from file:", loaded_books)

# 7. Pickle for serialization
def save_books_pickle(filename):
    with open(filename, 'wb') as f:
        pickle.dump(books, f)

def load_books_pickle(filename):
    # Open the file in binary read mode and deserialize the list using pickle
    try:
        with open(filename, 'rb') as f:
            return pickle.load(f)
    except FileNotFoundError:
        print(f"Error: {filename} not found.")
        return []
    except pickle.UnpicklingError:
        print(f"Error: {filename} could not be unpickled (corrupted or invalid format).")
        return []

# Test pickle functions
save_books_pickle("books.pkl")
print("Books loaded from pickle:", load_books_pickle("books.pkl"))

# 8. SQLite Database Interaction
def create_books_table():
    conn = sqlite3.connect('library.db')
    cursor = conn.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS books
                      (id INTEGER PRIMARY KEY, title TEXT, quantity INTEGER)''')
    conn.commit()
    conn.close()

def add_book_to_db(title, quantity):
    # Insert a new book record into the books table
    try:
        conn = sqlite3.connect('library.db')
        cursor = conn.cursor()
        cursor.execute("INSERT INTO books (title, quantity) VALUES (?, ?)", (title, quantity))
        conn.commit()
        print(f"Added '{title}' (quantity: {quantity}) to the database.")
    except sqlite3.Error as e:
        print(f"Database error: {e}")
    finally:
        conn.close()

# Test database functions
create_books_table()
add_book_to_db("Python for Beginners", 10)

# Verify the book was added by querying the table
conn = sqlite3.connect('library.db')
cursor = conn.cursor()
cursor.execute("SELECT * FROM books")
print("Books in database:", cursor.fetchall())
conn.close()


# 9. Web API Interaction (GitHub API)
def get_python_repos():
    url = "https://api.github.com/search/repositories"
    params = {"q": "language:python", "sort": "stars", "order": "desc"}
    response = requests.get(url, params=params)
    data = response.json()
    return [repo['name'] for repo in data['items'][:5]]

# Test API function
print("Top 5 Python repositories on GitHub:", get_python_repos())

print("\nLibrary Management System operations completed.")

# 10. Bonus Enhancement: Remove a book from the library
def remove_book(book_name):
    # Check if the book exists before trying to remove it, to avoid errors
    if book_name in books:
        books.remove(book_name)
        print(f"'{book_name}' has been removed from the library.")
        return True
    else:
        print(f"'{book_name}' not found in the library — nothing to remove.")
        return False

# Test the remove_book function
remove_book("Clean Code")          # should succeed
remove_book("Nonexistent Book")    # should fail gracefully
print("Book list after removal:", books)

```
