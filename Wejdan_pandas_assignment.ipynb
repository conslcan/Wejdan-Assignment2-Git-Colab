```python
# Task 5: Concatenation and Merging
# Concatenate two DataFrames vertically and merge them horizontally

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

# Your code here:
# Concatenate df1 and df2 vertically (stacking rows on top of each other)
df_concat = pd.concat([df1, df2], ignore_index=True)
print("Concatenated (vertical):\n", df_concat)

# Merge the concatenated result with df3 horizontally (side by side, matching by row position)
df_merged = pd.merge(df_concat, df3, left_index=True, right_index=True)
print("\nMerged (horizontal):\n", df_merged)

print("\nShape of final merged DataFrame:", df_merged.shape)
print("\nNote: merge() with left_index/right_index=True defaults to an inner join,")
print("so only rows with matching indices in both DataFrames are kept.")
print("df3 only has 2 rows (index 0-1), so rows with index 2-3 from df_concat are dropped.")
```

```python
# Task 8: Functions and Lambda
# Create a function to calculate the square of a number.
# Then, use a lambda function to do the same.

# Your code here:
# Regular function
def square(n):
    return n ** 2

# Lambda function (equivalent, written as a single expression)
square_lambda = lambda n: n ** 2

# Demonstrate both with two different inputs
print("Regular function: square(4) =", square(4))
print("Regular function: square(7) =", square(7))
print("Lambda function: square_lambda(4) =", square_lambda(4))
print("Lambda function: square_lambda(7) =", square_lambda(7))
```

```python
# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:
# Iterator: a class that implements __iter__ and __next__
class EvenNumbersIterator:
    def __init__(self, count):
        self.count = count
        self.current = 0
        self.num_yielded = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.num_yielded >= self.count:
            raise StopIteration
        value = self.current
        self.current += 2
        self.num_yielded += 1
        return value

# Demonstrate the iterator
print("Using the iterator class:")
even_iterator = EvenNumbersIterator(5)
for num in even_iterator:
    print(num)

# Generator: a function that uses 'yield' instead of 'return'
def even_numbers_generator(count):
    current = 0
    yielded = 0
    while yielded < count:
        yield current
        current += 2
        yielded += 1

# Demonstrate the generator
print("\nUsing the generator function:")
for num in even_numbers_generator(5):
    print(num)
```

```python
# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

# Your code here:
# map: applies a function to every element, returns a new sequence
squared = list(map(lambda x: x ** 2, numbers))
print("Squared numbers (map):", squared)

# reduce: repeatedly combines elements into a single cumulative value
product = reduce(lambda x, y: x * y, numbers)
print("Product of all numbers (reduce):", product)

# filter: keeps only elements where the function returns True
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers (filter):", evens)
```

```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Create at least two Rectangle objects
rect1 = Rectangle(5, 3)
rect2 = Rectangle(10, 7)

print(f"Rectangle 1: length={rect1.length}, width={rect1.width}")
print(f"Rectangle 1 area: {rect1.area()}")
print(f"Rectangle 1 perimeter: {rect1.perimeter()}")

print(f"\nRectangle 2: length={rect2.length}, width={rect2.width}")
print(f"Rectangle 2 area: {rect2.area()}")
print(f"Rectangle 2 perimeter: {rect2.perimeter()}")
```

```python
# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# Your code here:
# 1. Average salary by department
avg_salary_by_dept = df_employees.groupby('Department')['Salary'].mean()
print("Average salary by department:\n", avg_salary_by_dept)

# 2. Names of employees with salary > 60000
high_earners = df_employees[df_employees['Salary'] > 60000]['Name']
print("\nEmployees with salary > 60000:\n", high_earners.tolist())

# 3. Add a 'Bonus' column (10% of salary)
df_employees['Bonus'] = df_employees['Salary'] * 0.10
print("\nDataFrame with Bonus column:\n", df_employees)
```
